```python
import csv
```

```python
f = open('environmental.csv', 'rb')
reader = csv.reader(f)
temperature = []
wind = []
rownum = 0
for row in reader:
    if rownum == 0:
        c = row
    else:
        v = map(float, row[2:])
        temperature.append(v[0])
        wind.append(v[1])
    
    rownum += 1

print temperature
print wind
    
f.close()
```

```python
from pylab import *
```

```python

scatter(temperature, wind)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('atmosphere temperature and wind ')
ax.set_xlabel('atmosphere temperature')
ax.set_ylabel('atmosphere wind')
ax.scatter(temperature, wind)
fig.savefig("temp_wind.png")
```
